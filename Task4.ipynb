```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv("USvideos.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.drop_duplicates()
df.shape
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Removing columns whose values are in links/ not required
columns_to_removed = ['thumbnail_link','description']
df = df.drop(columns=columns_to_removed)
df.info()
```

```python
import datetime
```

```python
# Standardized the date and time format for trending_date column 
df['trending_date'] = df['trending_date'].apply(lambda x: datetime.datetime.strptime(x,'%y.%d.%m'))
df.head()
```

```python
# Standardized the date and time format for publish_time column 
df['publish_time'] = pd.to_datetime(df['publish_time'])
df.head()
```

```python
# Creating new columns for date and time insights
df['publish_month'] = df['publish_time'].dt.month
df['publish_day'] = df['publish_time'].dt.day
df['publish_hour'] = df['publish_time'].dt.hour
df.head()
```

```python
# Sorting data w.r.t cateogory_id
print(sorted(df['category_id'].unique()))
[1, 2, 10, 15, 17, 19, 20, 22, 23, 24, 25, 26, 27, 28, 29, 30, 43]
```

```python
# Mapping Category Ids to Category Names
df['category_name'] = np.nan
df.loc[(df["category_id"] == 1), "category_name"] = 'Film and Animation'
df.loc[(df["category_id"] == 2), "category_name"] = 'Autos and Vehicle'
df.loc[(df["category_id"] == 10), "category_name"] = 'Music'
df.loc[(df["category_id"] == 15), "category_name"] = 'Poets and Animals'
df.loc[(df["category_id"] == 17), "category_name"] = 'Sports'
df.loc[(df["category_id"] == 19), "category_name"] = 'Travel and Events'
df.loc[(df["category_id"] == 20), "category_name"] = 'Gaming'
df.loc[(df["category_id"] == 22), "category_name"] = 'People and Blogs'
df.loc[(df["category_id"] == 23), "category_name"] = 'Comedy'
df.loc[(df["category_id"] == 24), "category_name"] = 'Entertainment'
df.loc[(df["category_id"] == 25), "category_name"] = 'News and Politics'
df.loc[(df["category_id"] == 26), "category_name"] = 'Food and Travel'
df.loc[(df["category_id"] == 27), "category_name"] = 'Fashion and Lifestyle'
df.loc[(df["category_id"] == 28), "category_name"] = 'Science and Technology'
df.loc[(df["category_id"] == 29), "category_name"] = 'Education'
df.loc[(df["category_id"] == 30), "category_name"] = 'Business and Investing'
df.loc[(df["category_id"] == 43), "category_name"] = 'House Tour and Guide'
df.head()
df.shape
```

```python
# Visualization yearly count of published videos
df["year"] = df['publish_time'].dt.year
yearly_counts = df.groupby('year')['video_id'].count()

# Creating a bar chart
yearly_counts.plot(kind='bar', xlabel='Year', ylabel='Total Publish Count', title="Total Publish Video Per Year")

# Displaying chart
plt.show()
```

```python
# Visualizing Yearly Total Views
yearly_views = df.groupby('year')['views'].sum().reset_index()

#Bar Chart
yearly_counts.plot(kind='bar', xlabel='Year', ylabel='Total Views Count', title="Total Views Per Year")
plt.xticks(rotation=0)
plt.tight_layout()

# Displaying
# plt.bar(yearly_views['year'],yearly_views['views'])
# plt.xlabel('Year', fontsize=12)
# plt.ylabel('Total Views Count', fontsize=12)
# plt.title('Total Views Per Year', fontsize=15)
# plt.tight_layout()
# plt.show()
```

```python
# Visualizing Top 5 Categories by Total Views
category_views = df.groupby('category_name')['views'].sum().reset_index()

# sort the categories by views in descending order
top_categories = category_views.sort_values(by='views', ascending=False).head(5)

# Bar Chart 
plt.bar(top_categories['category_name'],top_categories['views'])
plt.xlabel('Category Name', fontsize=12)
plt.ylabel('Total Views', fontsize=12)
plt.title('Top 5 categories', fontsize=15)
plt.tight_layout()
plt.show()
```

```python
# Visualizing category-wise video count 
plt.figure(figsize=(12, 6))
sns.countplot(x='category_name', data=df, order=df['category_name'].value_counts().index)
plt.xticks(rotation = 90)
plt.title('Video count by Category')
plt.show()
```

```python
# Visualizing count of videos published per hour
videos_per_hour = df['publish_hour'].value_counts().sort_index()
# print(videos_per_hour)

plt.figure(figsize=(12, 6))
sns.barplot(x = videos_per_hour.index, y = videos_per_hour.values, palette='rocket')
plt.title('Number of videos published per hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of videos')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualizing trend of videos published over time

df['publish_time'] = pd.to_datetime(df['publish_time'])
df['publish_date'] = df['publish_time'].dt.date

videocount_by_date = df.groupby('publish_date').size()

plt.figure(figsize=(12, 6))
sns.lineplot(data=videocount_by_date)
plt.title('Videos Published Over Time', fontsize=14)
plt.xlabel('Publish Date', fontsize=14)
plt.ylabel('Number of videos', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot of Views vs. Likes

sns.scatterplot(data=df, x='views', y='likes')
plt.title('Views vs Likes')
plt.xlabel('Views')
plt.ylabel('Likes')
plt.show()
```

```python
# Analyzing Disabled Comments, Ratings, and Error/Removed Videos

plt.figure(figsize = (14, 8))
plt.subplots_adjust(wspace = 0.2, hspace = 0.4, top = 0.9)

plt.subplot(2,2,1)
g = sns.countplot(x='comments_disabled', data =df)
g.set_title('Comments Disabled', fontsize=16)

plt.subplot(2, 2, 2)
g1 = sns.countplot(x='ratings_disabled', data =df)
g1.set_title('Ratings Disabled', fontsize=16)

plt.subplot(2,2,3)
g2 = sns.countplot(x='video_error_or_removed', data =df)
g2.set_title('Video Error Or Removed', fontsize=16)

plt.show()
```
