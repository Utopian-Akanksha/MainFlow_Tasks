```python
questions = [
    "1. How many people are suffering with and without heart disease?",
    "2. Which sex has the most heart disease?",
    "3. People of which sex experience which type of chest pain most?",
    "4. People with which chest pain type are most prone to have heart disease?"
]
questions
```

```python
# 1. How many people are suffering with and without heart disease?
print(df['target'].value_counts())

#Plotting Bar chart
df.target.value_counts().plot( kind = 'bar', color = ['orchid', 'blue'])
plt.title("Heart Disease Prevalence in the Dataset")
plt.xlabel(" Heart Disease Status (1 = Present, 0 = Absent)")
plt.ylabel("Number of Individuals")
plt.xticks(rotation = 0)
plt.show()
```

```python
# No. of male and female in the dataset
df.sex.value_counts()
```

```python
#Plotting Pie chart
df.sex.value_counts().plot( kind = 'pie', figsize = (6, 4))
plt.title("Male-Female Ratio")
plt.legend(["Male", "Female"]);
```

```python
# 2. Which sex has the most heart disease?
print(pd.crosstab( df.target, df.sex ))

sns.countplot(x='target', data=df, hue='sex', palette='pastel')  # Added a custom palette for better aesthetics
plt.title("Heart Disease Frequency by Sex")  # More concise and clear title
plt.xlabel("Heart Disease Status (1 = With Disease, 0 = Without Disease)")  # Improved x-axis label
plt.ylabel("Number of Individuals")  # Added y-axis label for clarity
plt.legend(title="Sex", labels=["Female (0)", "Male (1)"])  # Customized legend for better understanding
plt.show()
```

```python
# 3. People of which sex experience which type of chest pain most?
print(df.cp.value_counts())

fig, axes = plt.subplots(1, 2, figsize=(14, 6))  # 1 row, 2 columns

# Plot 1: Chest Pain Type Distribution
df['cp'].value_counts().plot(kind='bar', color=['salmon', 'lightskyblue', 'springgreen', 'khaki'], ax=axes[0])
axes[0].set_title('Chest Pain Type Distribution')
axes[0].set_xlabel('Chest Pain Type')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=0)

# Plot 2: Type of Chest Pain for Each Sex
pd.crosstab(df['sex'], df['cp']).plot(kind='bar', color=['coral', 'lightskyblue', 'plum', 'khaki'], ax=axes[1])
axes[1].set_title('Chest Pain Type by Sex')
axes[1].set_xlabel('Sex (0 = Female, 1 = Male)')
axes[1].set_ylabel('Count')
axes[1].tick_params(axis='x', rotation=0)

# Adjust layout for better readability
plt.tight_layout()
plt.show()
```

```python
# Question 3 - Refined Approach
print(pd.crosstab( df.sex, df.cp ))
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='cp', hue='sex', palette='Set2')  # Grouped bar chart
plt.title("Chest Pain Type Distribution by Sex", fontsize=14)
plt.xlabel("Chest Pain Type (0 = Typical Angina, 1 = Atypical Angina, 2 = Non-Anginal Pain, 3 = Asymptomatic)", fontsize=12)
plt.ylabel("Number of Individuals", fontsize=12)
plt.legend(title="Sex", labels=["Female (0)", "Male (1)"], fontsize=10)
plt.xticks(ticks=[0, 1, 2, 3], labels=["Typical Angina", "Atypical Angina", "Non-Anginal", "Asymptomatic"], rotation=0)
plt.show()
```

```python
# 4. People with which chest pain type are most prone to have heart disease?
print(pd.crosstab( df.cp, df.target))

# Calculate the proportion of people with heart disease for each chest pain type
chest_pain_heart_disease = (
    df.groupby(['cp', 'target'])['target']
    .count()
    .unstack(fill_value=0)
    .apply(lambda x: x / x.sum(), axis=1)  # Normalize to get proportions
)

# Plotting a bar chart for proportions
chest_pain_heart_disease[1].plot(kind='bar', color='teal', figsize=(10, 6))
plt.title("Proportion of Heart Disease by Chest Pain Type", fontsize=14)
plt.xlabel("Chest Pain Type (0 = Typical Angina, 1 = Atypical Angina, 2 = Non-Anginal Pain, 3 = Asymptomatic)", fontsize=10)
plt.ylabel("Proportion with Heart Disease", fontsize=12)
plt.xticks(ticks=[0, 1, 2, 3], labels=["Typical Angina", "Atypical Angina", "Non-Anginal", "Asymptomatic"], rotation=0)
plt.ylim(0, 1)  # Set y-axis limit to represent proportions
plt.show()
```

```python
# 5. Which age group is most affected by heart disease?

# Create age bins
df['age_group'] = pd.cut(df['age'], bins=range(20, 90, 10), right=False)

# Count heart disease cases in each age group
age_heart_disease = df[df['target'] == 1]['age_group'].value_counts().sort_index()

# Plotting
age_heart_disease.plot(kind='bar', color='crimson', figsize=(10, 6))
plt.title("Heart Disease Cases by Age Group", fontsize=14)
plt.xlabel("Age Group (in years)", fontsize=12)
plt.ylabel("Number of Individuals with Heart Disease", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. What is the correlation between cholesterol levels and heart disease?
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='chol', y='target', hue='target', palette='coolwarm')
plt.title("Cholesterol Levels vs Heart Disease", fontsize=14)
plt.xlabel("Cholesterol Level (mg/dL)", fontsize=12)
plt.ylabel("Heart Disease (0 = No, 1 = Yes)", fontsize=12)
plt.legend(title="Heart Disease")
plt.show()
```

```python
# 7. Is there a relationship between maximum heart rate and heart disease?
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='target', y='thalach', palette='Set3')
plt.title("Maximum Heart Rate vs Heart Disease", fontsize=14)
plt.xlabel("Heart Disease (0 = No, 1 = Yes)", fontsize=12)
plt.ylabel("Maximum Heart Rate (thalach)", fontsize=12)
plt.show()
```

```python
# 8. Does fasting blood sugar affect heart disease?
pd.crosstab(df['fbs'], df['target']).plot(kind='bar', color=['salmon', 'teal'], figsize=(10, 6))
plt.title("Heart Disease Cases by Fasting Blood Sugar Level", fontsize=14)
plt.xlabel("Fasting Blood Sugar (0 = <120 mg/dL, 1 = >120 mg/dL)", fontsize=12)
plt.ylabel("Number of Individuals", fontsize=12)
plt.legend(["No Heart Disease", "Heart Disease"], title="Heart Disease")
plt.xticks(rotation=0)
plt.show()
```

```python
# 9. What is the impact of resting blood pressure on heart disease?
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='target', y='trestbps', 
            palette='viridis')
plt.title("Resting Blood Pressure vs Heart Disease", fontsize=14)
plt.xlabel("Heart Disease (0 = No, 1 = Yes)", fontsize=12)
plt.ylabel("Resting Blood Pressure (mmHg)", fontsize=12)
plt.show()
```

```python
# 10. How does exercise-induced angina relate to heart disease?
pd.crosstab(df['exang'], df['target']).plot(kind='bar', color=['lightcoral', 'mediumseagreen'], figsize=(10, 6))
plt.title("Exercise-Induced Angina vs Heart Disease", fontsize=14)
plt.xlabel("Exercise-Induced Angina (0 = No, 1 = Yes)", fontsize=12)
plt.ylabel("Number of Individuals", fontsize=12)
plt.legend(["No Heart Disease", "Heart Disease"], title="Heart Disease")
plt.xticks(rotation=0)
plt.show()
```
