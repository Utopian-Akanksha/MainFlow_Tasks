```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("householdtask3.csv")
```

```python
data.head(10)
```

```python
# Scatter plot with year vs own
plt.scatter(data["year"],data["own"])

# Adding title to the plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel("YEAR")
plt.ylabel("OWN")

# Displaying the chart
plt.show()
```

```python
# Line Chart with year vs own
plt.plot(data["year"])
plt.plot(data["own"])

# Adding title to the plot
plt.title("Line Chart")

# Setting the X and Y labels
plt.xlabel("YEAR")
plt.ylabel("OWN")

# Displaying the chart
plt.show()
```

```python
# Bar Chart with year vs own
plt.bar(data["year"],data["own"])

# Adding title to the plot
plt.title("Bar Chart")

# Setting the X and Y labels
plt.xlabel("YEAR")
plt.ylabel("OWN")

# Displaying the chart
plt.show()
```

```python
# Histogram 
plt.hist(data["income"])

# Adding title to the plot
plt.title("Histogram")

# Setting the X and Y labels
plt.xlabel("INCOME")
# plt.ylabel("OWN")

# Displaying the chart
plt.show()
```
